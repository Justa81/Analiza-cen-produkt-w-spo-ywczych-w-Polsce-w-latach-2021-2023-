# 2.  Połączenie trzech powyższych DataFrame za lata 2021, 2022, 2023 .

```python
import pandas as pd

# Funkcja do połączenia danych z różnych lat w jeden DataFrame
def Data_summary_for_the_year(dataframes):
    """
    Łączy dane o cenach produktów z różnych lat w jeden DataFrame,
    zachowując ceny z każdego roku i województwa.
    
    :param dataframes: Słownik z DataFrame'ami dla różnych lat.
    :return: DataFrame z zestawieniem cen produktów w różnych latach i województwach.
    """
    # Lista do przechowywania DataFrame'ów połączonych na podstawie lat
    df_list = []

    # Iterujemy przez wszystkie DataFrame'y 
    for year, df in dataframes.items():
        # Dodajemy nową kolumnę z rokiem, aby rozróżnić dane
        df_year = df.copy()
        df_year['Rok'] = year
        
        # Dodajemy do listy DataFrame z roku
        df_list.append(df_year)

    # Łączymy wszystkie DataFrame'y w jeden
    df_all_year = pd.concat(df_list, axis=0, ignore_index=True)
    
    # Zwracamy połączony DataFrame
    return df_all_year

# Słownik z DataFrame'ami dla różnych lat
dataframes = {
    2021: prices_2021_recipe,
    2022: prices_2022_recipe,
    2023: prices_2023_recipe
}

# Uzyskanie ostatecznego zestawienia danych
df_summery_all_years = Data_summary_for_the_year(dataframes)
```

# 4. Analiza cenowa wybranych produktów spożywczych w latach 2021, 2022, 2023.
---

```python
# Składniki i ich ilości w kilogramach/litrach/sztukach
ingredients = {
    "masło": 1.0,       # opakowanie
    "cukier":0.2,        #kg
    "mleko": 0.1,       # l
    "śmietana": 0.5,    # opakowania
    "mąka pszenna": 0.25,  # kg
    "jaja": 4,          # sztuki
    "dżem": 1.38         # g (załóżmy pół szklanki to 0.5 kg)
}
# Mapowanie składników do nazw w tabeli
price_mapping = {
    "masło": "masło świeże o zawartości tłuszczu ok. 82,5% - za 200g",
    "cukier": "cukier biały kryształ - za 1kg",
    "mleko": "mleko krowie spożywcze o zawartości tłuszczu 3-3,5%, sterylizowane - za 1l",
    "śmietana": "śmietana 18% tłuszczu - za 200g",
    "mąka pszenna": "mąka pszenna - za 1kg",
    "jaja": "jaja kurze świeże - za 1szt.",
    "dżem": "dżem - za 360g"
}
```

```python
# Funkcja do obliczania kosztów składników na ciasto dla danego roku i województwa
def Calculate_the_cost_of_ingredients(prices, ingredients, price_mapping):
    """
    Funkcja oblicza koszt składników na ciasto w danym roku, dla danego województwa.
    :param prices: Ceny w danym roku dla danego województwa
    :param ingredients: Słownik zawierający ilości składników
    :param price_mapping: Mapowanie nazw składników
    :return: Całkowity koszt składników na ciasto
    """
    total_cost = 0
    for ingredient, quantity in ingredients.items():
        name_in_table = price_mapping.get(ingredient)  # Pobieramy nazwę składnika w tabeli
        if name_in_table:  # Sprawdzamy, czy istnieje w mapowaniu
            price_per_unit = prices.get(name_in_table)  # Pobieramy cenę składnika
            if price_per_unit:
                if ingredient == "jaja kurze świeże - za 1szt.":  # Przeliczenie ceny jajek na sztuki
                    cost = quantity * price_per_unit
                else:
                    cost = quantity * price_per_unit  # Przeliczenie ceny na odpowiednią jednostkę (np. za kg)
                total_cost += cost
    return total_cost
```

```python
# Funkcja do obliczania kosztów dla wszystkich województw i lat
def Cost_summary(prices_2021_recipe, prices_2022_recipe, prices_2023_recipe, ingredients, price_mapping):
    """
    Funkcja generuje zestawienie kosztów składników na ciasto dla każdego województwa i roku.
    :param prices_2021, prices_2022, prices_2023: DataFrame z cenami dla lat 2021, 2022, 2023
    :param ingredients: Słownik zawierający ilości składników
    :param price_mapping: Mapowanie nazw składników
    :return: DataFrame z kosztami dla każdego województwa i roku
    """
    # Lista województw
    voivodeships = prices_2021_recipe["Nazwa"].unique()
    
    # Słownik do przechowywania wyników
    total_costs = {"Województwa": [], "Koszt 2021 rok": [], "Koszt 2022 rok": [], "Koszt 2023 rok": []}

    # Iteracja po województwach
    for voivodeship in voivodeships:
        # Pobranie cen dla danego województwa w każdym roku
        prices_2021_voivodeship = prices_2021_recipe[prices_2021_recipe["Nazwa"] == voivodeship].iloc[0]
        prices_2022_voivodeship = prices_2022_recipe[prices_2022_recipe["Nazwa"] == voivodeship].iloc[0]
        prices_2023_voivodeship = prices_2023_recipe[prices_2023_recipe["Nazwa"] == voivodeship].iloc[0]
        
        # Obliczanie kosztów dla każdego roku
        total_cost_2021 = Calculate_the_cost_of_ingredients(prices_2021_voivodeship, ingredients, price_mapping)
        total_cost_2022 = Calculate_the_cost_of_ingredients(prices_2022_voivodeship, ingredients, price_mapping)
        total_cost_2023 = Calculate_the_cost_of_ingredients(prices_2023_voivodeship, ingredients, price_mapping)
        
        # Dodajemy wyniki do słownika
        total_costs["Województwa"].append(voivodeship)
        total_costs["Koszt 2021 rok"].append(total_cost_2021)
        total_costs["Koszt 2022 rok"].append(total_cost_2022)
        total_costs["Koszt 2023 rok"].append(total_cost_2023)
    
    # Tworzymy DataFrame z wynikami
    return pd.DataFrame(total_costs)

# Zestawienie kosztów dla wszystkich województw i lat
df_cost = Cost_summary(prices_2021_recipe, prices_2022_recipe, prices_2023_recipe, ingredients, price_mapping)
```

# 9 Wskaźnik cen towarów i usług konsumpcyjnych -inflacja


```python
inflation_2021 = pd.read_excel('Inflacja 2021.xlsx')
inflation_2022 = pd.read_excel('Inflacja 2022.xlsx')
inflation_2023 = pd.read_excel('Inflacja 2023.xlsx')
```

```python


# Funkcja do połączenia danych z różnych lat w jeden DataFrame
def inflation_total(dataframes):
    """
    Łączy dane o inflacji grup produktów z różnych lat w jeden DataFrame,
    zachowując ceny z każdego roku i województwa.
    
    :param dataframes: Słownik z DataFrame'ami dla różnych lat.
    :return: DataFrame z zestawieniem inflacji w różnych latach i województwach.
    """
    # Lista do przechowywania DataFrame'ów połączonych na podstawie lat
    df_list = []

    # Iterujemy przez wszystkie DataFrame'y (z różnych lat)
    for year, df in dataframes.items():
        # Dodajemy nową kolumnę z rokiem, aby rozróżnić dane
        df_year = df.copy()
        df_year['Rok'] = year
        
        # Dodajemy do listy DataFrame z roku
        df_list.append(df_year)

    # Łączymy wszystkie DataFrame'y w jeden
    df_all_year = pd.concat(df_list, axis=0, ignore_index=True)
    
    # Zwracamy połączony DataFrame
    return df_all_year

# Słownik z DataFrame'ami dla różnych lat
dataframes = {
    2021: inflation_2021,
    2022: inflation_2022,
    2023: inflation_2023
}

# Uzyskanie ostatecznego zestawienia danych
df_summery_inflation_years = inflation_total(dataframes)
```

# Porównanie wzrostu cen składników na ciasto do inflacji

```python
mean_cost_2021 = df_cost['Koszt 2021 rok'].mean()
mean_cost_2022 = df_cost['Koszt 2022 rok'].mean()
mean_cost_2023 = df_cost['Koszt 2023 rok'].mean()
```

```python
print(mean_cost_2021)
print(mean_cost_2022)
print(mean_cost_2023)
```

```python
# Obliczanie wzrostu procentowego średnich kosztów
growth_2021_2022 = ((mean_cost_2022 - mean_cost_2021) / mean_cost_2021) * 100
growth_2022_2023 = ((mean_cost_2023 - mean_cost_2022) / mean_cost_2022) * 100

# Zakładając inflację ogólną (można podać dane na podstawie rzeczywistej inflacji w danym roku)
inflation_2021_2022 = 14.4  # Inflacja ogólna w 2022 roku
inflation_2022_2023 = 11.4  # Inflacja ogólna w 2023 roku

# Porównanie wzrostu z inflacją
print(f"Średni koszt 2021: {mean_cost_2021:.2f} PLN")
print(f"Średni koszt 2022: {mean_cost_2022:.2f} PLN")
print(f"Średni koszt 2023: {mean_cost_2023:.2f} PLN")

print(f"Wzrost kosztów 2021-2022: {growth_2021_2022:.2f}%")
print(f"Wzrost kosztów 2022-2023: {growth_2022_2023:.2f}%")

print(f"Inflacja ogólna 2021-2022: {inflation_2021_2022:.2f}%")
print(f"Inflacja ogólna 2022-2023: {inflation_2022_2023:.2f}%")
```

# Wzrost kosztów z 2021 na 2022 wynosił 24.01%, co jest istotną zmianą. Wzrost ten jest wyższy niż inflacja ogólna (14.40%) w tym samym okresie, co sugeruje, że ceny produktów w badanej grupie rosły szybciej niż średnia inflacja w Polsce. Wzrost kosztów z 2022 na 2023 wynosił 9.94%, co jest już znacznie mniejszym wzrostem w porównaniu do poprzedniego roku. To wskazuje na pewną stabilizację cen w 2023 roku, chociaż nadal występuje wzrost kosztów, ale w mniejszym stopniu niż w 2022 roku.

```python
labels = ['2021-2022', '2022-2023']
wzrost_kosztow = [24.01, 9.94]  # Wzrost kosztów
inflacja = [14.40, 11.40]  # Inflacja ogólna

# Ustawienie pozycji na osi X dla słupków
x = np.arange(len(labels))

# Szerokość słupków
width = 0.35

# Tworzenie wykresu
fig, ax = plt.subplots(figsize=(8, 6))

# Tworzenie dwóch zestawów słupków
bars1 = ax.bar(x - width / 2, wzrost_kosztow, width, label='Wzrost kosztów', color='skyblue')
bars2 = ax.bar(x + width / 2, inflacja, width, label='Inflacja ogólna', color='salmon')

# Dodanie etykiet, tytułu i legendy
ax.set_ylabel('Procent (%)')
ax.set_title('Porównanie wzrostu kosztów i inflacji ogólnej w latach 2021-2022 i 2022-2023')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Dodanie wartości nad słupkami
def add_labels(bars):
    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width() / 2, yval + 0.5, f'{yval}%', ha='center', va='bottom')

# Dodanie wartości do słupków
add_labels(bars1)
add_labels(bars2)

# Wyświetlanie wykresu
plt.tight_layout()
plt.show()
```

Podczas gdy ogólny wskaźnik inflacji w 2022 roku wyniósł 14,4%, średnia cena składników do przygotowania ciasta wzrosła aż o 24%. Należy pamietać jednak że składniki które były analizowane to tylko malutka część koszyka inflacyjnego.W latach 2022-2023 ta różnica jest już mniejsza.
